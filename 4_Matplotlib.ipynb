```python
import csv
import matplotlib.pyplot as plt

f = open('Daejeon_temp.csv','r', encoding='cp949')
data = csv.reader(f)
header = next(data)
print(header)

high_temp = []
low_temp = []

for row in data:
    if row[-1]!='' and row[-2]!='':
        date = row[0].split('-')
        if date[1] == '02' and date[2] =='19':
            high_temp.append(float(row[-1]))
            low_temp.append(float(row[-2]))

f.close()

plt.title('2/19 Temperature chage')
plt.plot(high_temp,'red',label='high')
plt.plot(low_temp,'blue',label='low')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('Daejeon_temp.csv','r', encoding='cp949')
data = csv.reader(f)
header = next(data)
print(header)

Jan = []
Aug = []

for row in data:
    if row[-1] != '':
        date = row[0].split('-')
        if date[1] == '01':
            Jan.append(float(row[-1]))
        if date[1] == '08':
            Aug.append(float(row[-1]))
f.close()

print(max(Aug))
plt.title('Daejeon January and August Highest Temperature Frequency')
plt.hist(Jan,color='b', bins = 10, label = 'Jan')
plt.hist(Aug,color='r', bins = 10, label = 'Aug')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('Daejeon_temp.csv','r', encoding='cp949')
data = csv.reader(f)
header = next(data)
print(header)

avg_temp = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[2]!='':
        date = row[0].split('-')
        avg_temp[int(date[1])-1].append(float(row[2]))

f.close()

plt.title('Daejeon monthly average temperature distribution')
plt.boxplot(avg_temp)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('Daejeon_temp.csv','r', encoding='cp949')
data = csv.reader(f)
header = next(data)
print(header)

avg_temp = []
for i in range(31):
    avg_temp.append([])

for row in data:
    if row[2]!='':
        date = row[0].split('-')
        if date[1] == '08':
            avg_temp[int(date[2])-1].append(float(row[2]))

f.close()

plt.style.use('ggplot')
plt.figure(figsize=(10,5),dpi=300)
plt.title('Daejeon daily average temperature distribution',fontdict={'fontsize':30, 'fontweight':'light'})
plt.boxplot(avg_temp,showfliers=False)
plt.show()


```
