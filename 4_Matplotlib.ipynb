```python
import matplotlib.pyplot as plt
import random
#한글폰트 rcparams
plt.style.use('ggplot')

result =[]
for i in range(13):
    result.append(random.randint(1,1000))
print(sorted(result))
plt.boxplot(result)
plt.show()
import numpy as np
result = np.array(result)
print('1/4: '+str(np.percentile(result,25)))
print('2/4: '+str(np.percentile(result,50)))
print('3/4: '+str(np.percentile(result,75)))


plt.barh(['aa','bb','cc','dd','ee,','ff,'],[1,2,3,5,6,7])
plt.show()


color = ['darkmagenta','deeppink','hotpink','pink']
size = [2441,2312,1031,1233]
label = ['A','B','AB','O']
plt.title('A blood type')
plt.pie(size,labels=label, autopct='%.3f%%', colors=color, explode=(0,0.2,0,0.1))
plt.legend(loc=10)
plt.show()

plt.scatter([1,2,3,4],[0.5,2,1,3], s=[30,60,90,120],c=range(4), cmap='RdBu', alpha=1)
plt.colorbar()
```
